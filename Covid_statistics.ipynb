```python
#Libraries:
import pandas as pd
import numpy as np
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings("ignore")
```

```python
# Initialize driver:
driver = webdriver.Edge()
driver.get("https://www.nytimes.com/interactive/2021/world/brazil-covid-cases.html")

# Press buttons:
accept_button = driver.find_element("class name", "css-1fzhd9j")
accept_button.click()
show_all_button = driver.find_element(By.XPATH, '//button[@class="showall"]')
driver.execute_script("arguments[0].click();", show_all_button)

# Close driver:
html = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(html, "html.parser")
all_covid = soup.findAll("tbody", class_="children")[0]
```

```python
column_names = ["CITY", "CASES", "CASES PER 100K", "14-DAY CHANGE", "DEATHS", "DEATHS PER 100K"]
data = [[] for i in range(len(column_names))]

table_rows = all_covid.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    for j in range(len(row)):
        data[j].append(row[j])

pd.DataFrame.from_dict(dict(zip(column_names, data)))
```
