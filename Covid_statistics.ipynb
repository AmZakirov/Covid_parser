```python
#Библиотеки
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver = webdriver.Edge('C:\\msedgedriver.exe')
driver.get("https://www.nytimes.com/interactive/2021/world/brazil-covid-cases.html")
html = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(html, "html.parser")
all_covid = soup.findAll("tbody", class_="children")
```

```python
def unique_elements(l1,l2):
    new_list = []
    for item in l1:
        if item not in l2:
            new_list.append(item)
    return new_list
a = all_covid[0].findAll("td", class_="name")
b = all_covid[0].findAll("td", class_="bignum cases show-mobile")
c = all_covid[0].findAll("td", class_="num cases show-mobile")
d = all_covid[0].findAll("td", class_="chart cases wider show-mobile")
e = unique_elements(all_covid[0].findAll("td", attrs={"class": "bignum"}),all_covid[0].findAll("td", class_="bignum cases show-mobile"))
f = all_covid[0].findAll("td", class_="num td-end")
result = [[a[i].text, b[i].text, c[i].text, d[i].text, e[i].text] for i in range(len(a))]

indices = []
data = []
for i in range(len(result)):
    result[i][1] = result[i][1].replace(',','')
    indices.append(result[i][0])
    data.append([result[i][1],result[i][2],result[i][3],result[i][4]])
result
```

```python
#Сохраение в датафрейм
df = pd.DataFrame(data = data , index = indices , columns = ['Cases','Per_100,000','14day_change','Deaths']) 
df.head(len(data))
```
